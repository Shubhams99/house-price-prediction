```python
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:, -1]
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:, -1]
```

#SPLITTING THE DATASET

```python


#spliting the dataset
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
xtrain=xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
xtest=xtest.reshape(-1,1)
ytest=ytest.reshape(-1,1)

```

DEFINE ERROR FUNCTION

```python
def error(m,x,c,t):
    N=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/(2*N)
```

DEFINE UPDATE FUNCTION

```python
def update(x,m,c,t,learning_rate):
    
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

define gradient descent function

```python

def gradient_descent(init_m,init_c,x,t,learning_rate,iterations,error_threshold):
    m=init_m
    c=init_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if(e<error_threshold):
            
            
            print("error less than threshold stopping gradient descent ")
            print(e)
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
        
                
        
                
        
    

   
           
        
        
```


initialise all the values and pass those to the gradient_descent function

```python
%%time
init_m=0.7
init_c=0.2
learning_rate=0.001
iterations=250
error_threshold=0.001
m,c,error_values,mc_values= gradient_descent(init_m,init_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```


TRAINING THE MODEL


```python

#define regression object
lm=LinearRegression()
```

PLOTTIG THE ERROR VALUES

```python
plt.plot(np.arange(len(mc_values)),error_values)
plt.ylabel("Error")
plt.xlabel("Iterations")
```

```python
#fitting the model
lm.fit(xtrain,ytrain)
```

```python
lm.coef_
```

MAKING PREDICTION

```python

predictions=lm.predict(xtest)
```

```python
#plotting
plt.scatter(ytest,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predictions')
```

```python
#import evaluation matrix
from sklearn import metrics
print("MSE:",metrics.mean_squared_error(ytest,predictions))
print("RMSE:",np.sqrt(metrics.mean_squared_error(ytest,predictions)))
```
